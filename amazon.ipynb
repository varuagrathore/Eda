Before making analyzing the data, it is important to clean and prepare data. The methods used to clean and prepare the data are as listed below:


Changing Data Types of Columns from object to Floats

Filling in Missing Information

Checking For Duplicate Rows

```python
data['discounted_price'] = data['discounted_price'].astype(str)
data['actual_price'] = data['actual_price'].astype(str)

data['discounted_price']=data['discounted_price'].str.replace('₹','')
data['discounted_price']=data['discounted_price'].str.replace(',','')
data['discounted_price']=data['discounted_price'].astype('float64')

data['actual_price']=data['actual_price'].str.replace('₹','')
data['actual_price']=data['actual_price'].str.replace(',','')
data['actual_price']=data['actual_price'].astype('float64')
```

```python
data['discounted_price']
```

```python
data['discount_percentage'] = data['discount_percentage'].astype(str)
data['discount_percentage'] = data['discount_percentage'].str.replace('%','').astype('float64')
data['discount_percentage'] = data['discount_percentage']/100
data['discount_percentage']
```

```python
data['rating'].value_counts()
```

```python
#data[data['rating']=='|']

data.query('rating=="|"')
```

```python
data['rating']=data['rating'].astype(str)
data['rating']=data['rating'].str.replace('|','4.0').astype('float64')
```

```python
data['rating_count']=data['rating_count'].str.replace(',','').astype('float64')
```

```python
category_split=data1['category'].str.split('|',expand=True)
```

```python
category_split = category_split.rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
```

```python
#Fixing Strings in the Category_1 Column

data1['category_1'] = data1['category_1'].str.replace('&', ' & ')
data1['category_1'] = data1['category_1'].str.replace('OfficeProducts', 'Office Products')
data1['category_1'] = data1['category_1'].str.replace('MusicalInstruments', 'Musical Instruments')
data1['category_1'] = data1['category_1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
data1['category_2'].value_counts()
```

```python
#Fixing Strings in the Category_1 Column
data1['category_2'] = data1['category_2'].str.replace(',', ', ')
data1['category_2'] = data1['category_2'].str.replace('&', ' & ')
data1['category_2'] = data1['category_2'].str.replace('HomeTheater', 'Home Theater')
data1['category_2'] = data1['category_2'].str.replace('HomeStorage', 'Home Storage')
data1['category_2'] = data1['category_2'].str.replace('CarAccessories', 'Car Accessories')
data1['category_2'] = data1['category_2'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
data1['category_2'] = data1['category_2'].str.replace('HomeImprovement', 'Home Improvement')
data1['category_2'] = data1['category_2'].str.replace('PowerAccessories', 'Power Accessories')
data1['category_2'] = data1['category_2'].str.replace('OfficeElectronics', 'Office Electronics')
data1['category_2'] = data1['category_2'].str.replace('CraftMaterials', 'Craft Materials')
data1['category_2'] = data1['category_2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
data1['category_2'] = data1['category_2'].str.replace('HomeAudio', 'Home Audio')
data1['category_2'] = data1['category_2'].str.replace('NetworkingDevices', 'Networking Devices ')
data1['category_2'] = data1['category_2'].str.replace('WearableTechnology', 'Wearable Technology ')
data1['category_2'] = data1['category_2'].str.replace('OfficePaperProducts', 'Office Paper Products ')
data1['category_2'] = data1['category_2'].str.replace('NetworkingDevices', 'Networking Devices ')
```

```python
rating_score = []

for score in data1['rating']:
    if score < 2.0 : rating_score.append('Poor')
    elif score < 3.0 : rating_score.append('Below Average')
    elif score < 4.0 : rating_score.append('Average')
    elif score < 5.0 : rating_score.append('Above Average')
    elif score == 5.0 : rating_score.append('Excellent')
```

```python

data1['rating_score'] = rating_score

data1['rating_score'] = data1['rating_score'].astype('category')
```

```python
data1['rating_score'] = data1['rating_score'].cat.reorder_categories(['Below Average', 'Average', 'Above Average', 'Excellent'], ordered=True)
```

```python
data1['difference_price'] = data1['actual_price'] - data1['discounted_price']
```
